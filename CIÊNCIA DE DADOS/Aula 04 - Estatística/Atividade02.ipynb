Importando e Carregando o arquivo .csv


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

notasAlunos = pd.read_csv('notas_alunos.csv')
notasAlunos.head()
```

Criando um Dataframe Notas

```python

dados = notasAlunos.Nota
dados.head()
```

Função para calcular as estatisticas requisitadas:

```python
def estatisticas(nome, dados):
    print(f"--- {nome} ---")
    print(f"Média: {np.mean(dados)}")
    print(f"Mediana: {np.median(dados)}")
    print(f"Moda: {pd.Series(dados).mode().tolist()}")
    print(f"Variância: {np.var(dados, ddof=1)}")
    print(f"Desvio Padrão: {np.std(dados, ddof=1)}")

estatisticas("Notas: ",dados)
```

Identificando maior e menor nota da turma


```python
# Identificar o(s) aluno(s) com a maior nota
maior = dados.max()
print("Maior nota:")
print(maior)

menor = dados.min()
print("Menor nota:")
print(menor)


```

```python
amplitude = maior - menor
print(f"Amplitude: {amplitude}")
```

Classificando Alunos pelas estatisticas obtidas:

```python
# Definir as faixas de notas
def classificar(nota):
    if 0 <= nota < 5:
        return "Insuficiente"
    elif 5 <= nota < 7:
        return "Regular"
    elif 7 <= nota < 9:
        return "Bom"
    elif 9 <= nota <= 10:
        return "Excelente"

# Aplicar a classificação às notas
notasAlunos["Faixa"] = dados.apply(classificar)

# Contar o número de alunos em cada faixa
contagem_faixas = notasAlunos["Faixa"].value_counts()

# Calcular a porcentagem de alunos em cada faixa
percentuais = (contagem_faixas / len(dados)) * 100

# Exibir os resultados
print("Contagem de alunos por faixa:")
print(contagem_faixas)

print("\nPorcentagem de alunos por faixa:")
print(percentuais)
```

```python
# 1. Histograma para mostrar a distribuição das notas
plt.hist(dados, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribuição das Notas')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Boxplot para visualizar a dispersão e identificar outliers
plt.boxplot(dados, vert=False, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title('Boxplot das Notas')
plt.xlabel('Notas')
plt.show()


# Gráfico de barras
plt.bar(contagem_faixas.index, contagem_faixas.values, color=['red', 'orange', 'blue', 'green'])
plt.title('Quantidade de Alunos por Faixa de Nota')
plt.xlabel('Faixa de Nota')
plt.ylabel('Quantidade de Alunos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Análise Crítica a partir dos dados

A média (6.925) e a mediana (7.0) estão próximas, indicando uma distribuição aproximadamente simétrica. Simetria dos Dados
A proximidade entre média (6.925) e mediana (7.0) confirma uma distribuição simétrica, sem grandes desvios para extremos.
