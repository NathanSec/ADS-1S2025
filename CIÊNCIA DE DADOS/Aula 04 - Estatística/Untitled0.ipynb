# **Exercício**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios de tempo de ciclo de uma máquina
tempos_ciclo = [12, 15, 13, 14, 16, 12, 17, 15, 14, 13]

media = np.mean(tempos_ciclo)
mediana = np.median(tempos_ciclo)
moda = pd.Series(tempos_ciclo).mode().tolist()
variancia = np.var(tempos_ciclo, ddof=1)
desvio_padrao = np.std(tempos_ciclo, ddof=1)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Variância: {variancia}")
print(f"Desvio Padrão: {desvio_padrao}")

#Histograma
plt.hist(tempos_ciclo, bins=15, edgecolor='black')
plt.title('Distribuição dos Tempos de Ciclo')
plt.xlabel('Tempo (min)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

maquinaA = [3, 2, 4, 5, 3, 4, 2, 3, 5, 4]
maquinaB = [6, 5, 7, 6, 5, 6, 7, 6, 5, 6]
maquinaC = [1, 3, 2, 4, 2, 3, 4, 2, 3, 2]

def estatisticas(nome, dados):
    print(f"--- {nome} ---")
    print(f"Média: {np.mean(dados)}")
    print(f"Mediana: {np.median(dados)}")
    print(f"Moda: {pd.Series(dados).mode().tolist()}")
    print(f"Variância: {np.var(dados, ddof=1)}")
    print(f"Desvio Padrão: {np.std(dados, ddof=1)}")

# Chamadas da função fora da definição da função
estatisticas("Máquina A", maquinaA)
estatisticas("Máquina B", maquinaB)
estatisticas("Máquina C", maquinaC)
```
